### GIC

```


P  -> começo LS final
LS -> S | S LS
S  -> ID ::= E; | ler(LI); | escrever(LE);
LI -> ID | ID , LI
LE -> E | E , LE
E  -> O | O OP E
O  -> ID | K | (E) 

```

### BNF
Utiliza metasímbolos como < > para no terminales , ::= para definiciones  Y |	unión de reglas

```
<programa>        ::= começo <listaSentencias> final
<listaSentencias> ::= <sentencia> | <sentencia> <listaSentencias>
<sentencia>       ::= <id> ::= <exp>; | ler(<listaID>); | escrever(<listaExp>);
<listaID>         ::= <id> | <id> , <listaID>
<listaExp>        ::= <exp> | <exp> , <listaExp>
<exp>             ::= <primaria> | <primaria> <opeAditivo> <exp>
<primaria>        ::= <id> | <cte> | (<exp>)  


```


### EBNF
Incorpora símbolos como:
```
[ ]	elemento optativo
( | )	selección de una alternativa
{ }	repetición
*	0 o más veces
+	1 o más veces
?	0 ó 1 vez
```
```
<programa>        ::= começo {<sentencias>}+ final
<sentencia>       ::= <id> ::= <exp>; | ler(<listaID>); | escrever(<listaExp>);
<listaID>         ::=  <id> {, <id>}*
<listaExp>        ::=  <exp> {,<exp> }*
<exp>             ::= <primaria> {<opeAditivo><primaria> }*   
<primaria>        ::= <id> | <cte> | (<exp>)  


```




### ABNF

Según el estilo descrito en las fuentes (estilo ANSI C), usa itálicas para no terminales, negritas para terminales y : para reglas.


 

*programa*: **começo** *sentencias* **final**

*sentencias*: *sentencia* { *sentencia* }

*sentencia*:  <sub>uno de</sub>´*identificador* := *expresión* ; **ler**(*identificadores*) ; **escrever**(expresiones) ;

*identificadores*:  *identificador* {**,**  *identificador* }

*expresiones*:  *expresión* {**,**  *expresión* }

*expresión*:  *primaria* { *operadorAditivo* *primaria*}

*primaria*:<sub>uno de</sub> *identificador*  { *expresión* }

*identificador*: *letra* { *letraODigito* }

*letraODigito*:  <sub>uno de</sub> *letra* *digito*

*letra*: <sub>uno de</sub> **a-z** **A-Z**

*digito*:<sub>uno de</sub> **0-9**

*operadorAditivo*: <sub>uno de</sub> **+** **-**
